```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import StandardScaler
```

```python
df = fetch_california_housing(as_frame=True)
type(df)
```

```python
x = df.data
y = df.target
```

```python
vit_data = pd.DataFrame()
vit_data['Features'] = x.columns
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
vit_data['VIF'] = [variance_inflation_factor(x_scaled,i) for i in range(x_scaled.shape[1])]
vit_data
```

```python
x.shape[1]
```

```python
x = x[['MedInc','HouseAge','AveRooms','Population','AveOccup','Latitude']]
x_scaled = scaler.fit_transform(x)
x_scaled
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled, y, test_size=0.2, random_state=2)
```

```python
x_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
model.intercept_
model.coef_
```

```python
y_pred =model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('Mearn Absolute Error is ',mean_absolute_error(y_test,y_pred))
print('Mearn Square Error is ',mean_squared_error(y_test,y_pred))
print('R2_Score is ',r2_score(y_test,y_pred))
```

#LASSO Rregression Algorithem

```python
from sklearn.linear_model import Lasso, Ridge, ElasticNet
```

```python
#Initialize Model
lasso_model = Lasso()
#Train model
lasso_model.fit(x_train, y_train)
#Predict the model
lasso_y_pred = lasso_model.predict(x_test)
print('lasso - Mearn Absolute Error is ',mean_absolute_error(y_test,lasso_y_pred))
print('lasso - Mearn Square Error is ',mean_squared_error(y_test,lasso_y_pred))
print('lasso - R2_Score is ',r2_score(y_test,lasso_y_pred))
```

#Ridge Rregression Algorithem

```python
#Initialize Model
ridge_model = Ridge()
#Train model
ridge_model.fit(x_train, y_train)
#Predict the model
ridge_y_pred = ridge_model.predict(x_test)
print('Ridge - Mearn Absolute Error is ',mean_absolute_error(y_test,ridge_y_pred ))
print('Ridge - Mearn Square Error is ',mean_squared_error(y_test,ridge_y_pred ))
print('Ridge - R2_Score is ',r2_score(y_test,ridge_y_pred ))
```
